Breaking down the names in the first example:

* **ggplot**: function to create and return a ggplot object. Note that the evaluating the object causes it to be plotted wherever R is currently plotting things; in Jupyter this will be an output cell.
* **geom_point**: function returns a gg geometry object. Note that the geometry is added to the ggplot object.
* **mapping, data, x, y**: parameter keys for ggplot, geom_point and aes function calls
* **aes**: function to create a gg aesthetic object.
* **mpg**: a data frame object included as sample data with the ggplot2 package
* **displ**, **hwy** and **class**: data columns in the mpg data frame

### <xnum>3.4.1</xnum> Aesthetics

We're first going to explore the aesthetic mapping components of Wickham's **grammer of graphics**. The aesthetic defines how the plot elements are to be represented, and there are many aesthetics you can modify:

set the size aesthetic use values from the mpg$class column.
And then try with color, shape and alpha (transparency)

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, size = class))
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, color = class))
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, shape = class))
```

```r
# ggplot2 only recommends less than 7 shapes... so you have to override the default
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, shape = class)) + scale_shape_manual(values = 15:22)
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, alpha = class))
```

### <xnum>3.4.2</xnum> Facets

ggplot2 uses the word **facet** for splitting the data into multiple plots.

Note the use of the ~ operator to create formula objects in the following 

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + 
  facet_wrap(~ class, nrow = 2)
```

facet_grid does a two variable x-y arrangement of facets sub-plots: 

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + 
  facet_grid(drv ~ cyl)
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(drv ~ .)
```

### <xnum>3.4.3</xnum> Geometry

The geom (or geometry) of a gg plot determines the structure of the plot visualization.  The mapping aesthetic has to make sense for the particular geometry being used. For example, the linetype aesthetic wouldn't make sense for a point geom; just as shape wouldn't make sense for a line or smooth geometry.


```r
ggplot(data = mpg) + 
  geom_smooth(mapping = aes(x = displ, y = hwy))
```

```r
ggplot(data = mpg) + 
  geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv))
```

You can layer your geoms on the same plot:

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  geom_smooth(mapping = aes(x = displ, y = hwy))
```

And you can have your aesthetic attached to the plot object so it applies as a default to all geom layers:

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point() + 
  geom_smooth()
```

And you can override or add additional features like data and aesthetics on a layer by layer basis:

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point(mapping = aes(color = class)) + 
  geom_smooth(data = filter(mpg, class == "subcompact"), se = FALSE)
```

### <xnum>3.4.4</xnum> Statistical Transformations

Transformations of data can be computed by R code before you do the plot, but often a particular transformation is  part of the concept of the plot, so ggplot incorporates this concept explicitly as part of plotting. Bar charts are a natural example since they are often based on counts or frequency in the data set, and here's the ggplot2::diamonds sample data set for the example:  

```r
head(diamonds)
```

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut))
```

The y-axis value for the bars is not in the diamonds data frame - it is a transformation of the data using stat_count() - the default stat transformation for geom_bar().

```r
?geom_bar
```

[Rbook]: https://r4ds.had.co.nz/
[WickhamAdvanced]: http://adv-r.had.co.nz/

[Wickham][Rbook] gives the following example of overriding the aes default y aesthetic chosen by the stat transformation and using the computed value *..prop..* instead.  This is a  bit obscure at this point, since there is no introduction by Wickham to computed values ("..prop..") or the use of the group aesthetic up to this point. If you check the documentation, you see that the proportion value is computed internatally, and setting the grouping statistic to a constant int instead of using a column name from the data frame indicates proportions should be counted against the entire data set, not just within specific groups. Group answers the question "proportion of what?"

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, y = ..prop.., group = 123))
```

Particular stats also have default geoms. In the next example, thet stat_summary uses its default pointrange geom.


```r
ggplot(data = diamonds) + 
  stat_summary(
    mapping = aes(x = cut, y = depth),
    fun.ymin = min,
    fun.ymax = max,
    fun.y = median
  )
```

Next we show that:
* You have to be careful with name confusion *class* is also a base function in R
* object members are selected with dollar ($).  Dot (.) in names is not an operation, but is conventionally used for name heirarchies or to indicate a class implementation of a generic function; a leading dot is used for name hiding (like leading underscore in Python)

`?class` or `help(class)` will work for R as its does for Python. 

```r
var1 <- ggplot(mpg, aes(displ, hwy, color=class)) + geom_point()
cat("the base type of var 1 is:"); typeof(var1)
cat("var1 is a member of the following classes: "); class(var1);
cat("var1 has the following names:  "); names(var1)
cat("the labels name within var 1 has the value:  "); var1$labels
cat("var1$labels is a: "); typeof(var1$labels)
cat("and also has names: "); names(var1$labels)
```

### <xnum>3.4.5</xnum>  Aside: R programming 'gotcha's

#### no magic!

The R kernel won't do magic in Jupyter notebooks.

#### non standard evaluation

In most programming languages like Python and Javascript, the previous example 

    ggplot(mpg, aes(displ, hwy, color=class)) + geom_point()

would be a problem.  The names displ and hwy are not in scope, so a call like `aes(displ, hwy, color=class)`  would cause an error when the interpreter tries to evaluate those symbols. They simply have no value.

R has a capability called NSE [**non-standard evaluation**](http://adv-r.had.co.nz/Computing-on-the-language.html), which means the parameters do not have to be evaluated when a function is called - they can be evaluated later, when the value is needed. For a quicker overview of NSE and realted R concepts of quoting and promises, you can also check out an [NSE discussion on Edwin Thoen's R blog](https://edwinth.github.io/blog/nse/).  This often makes understanding a line of code conflict with a seasoned programmer's expectations. 

<span class=notcovered>One of the challenges of using tidyverse and gglot2 id they rely heavily on NSE. If you want to pursue R programming in depth you'll need some understanding of the related concepts.<span>

#### objects and classes in R

Beyond the base object types, there are three different [object-oriented systems in R](http://adv-r.had.co.nz/OO-essentials.html), with different behaviours that have been introduced over time.  In systems S3 and S4, generic function calls are dispatched to specific functions based on the object invoked. S4 adds formal class definitions which are not present in S4. The RC system introduces more conventional message-passing, with messages attached to classes instead of being associated with generic functions. These distinctions become important when you write complex code in R to extend packages or write your own classes. For a lot of people using R for basic scripting, these details may remain unexplored.

Most R packages use S3.



### <xnum>3.4.6</xnum> Position adjustments

Many plots are fine if we let the data mark appear wherever it falls in the x-y coordinate system. But sometimes this does not convey the information we want.  For example, we can add a third value to a bar chart using fill color. First, let's look at the bar positions without adding a new value, just re-using the same y (based on the stat_count of cut):

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = cut))
```

But if we use a third data value for out plot, the position of the color bars will be adjusted so the bars don't overlap.

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = clarity))
```

Or we can override the default and go back to basing all the counts at 0 on the y axis. This is generally confusing, because it violates conventions people are expecting:

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = clarity), position="identity")
```

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut, fill = clarity), position="dodge")
```

Instead of dodge, jitter is a good option to prevent points from overlapping on a point plot:

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy), position = "jitter")
```

### <xnum>3.4.7</xnum> Co-ordinate systems

ggplot2 provides a few different co-ordinate systems.  They complicate matters because all the other constructs for a plot have to be interpreted differently if you're using different co-ordinates, and ggplot has more trouble picking good defaults. Here are a few examples from Wickham.

```r
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) + 
  geom_boxplot()
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) + 
  geom_boxplot() +
  coord_flip()
```

```r
bar <- ggplot(data = diamonds) + 
  geom_bar(
    mapping = aes(x = cut, fill = cut), 
    show.legend = FALSE,
    width = 1
  ) + 
  theme(aspect.ratio = 1) +
  labs(x = NULL, y = NULL)

bar + coord_flip()
bar + coord_polar()
```

[R4DS](https://r4ds.had.co.nz) chapter 3 doesn't discuss how to do titles or labels on R plots, but the labs function is pretty straight-forward:

```r
?labs
```

```r
p <- ggplot(mtcars, aes(mpg, wt, colour = cyl)) + geom_point()
p + labs(colour = "Cylinders")
p + labs(x = "New x label")

# The plot title appears at the top-left, with the subtitle
# display in smaller text underneath it
p + labs(title = "New plot title")
p + labs(title = "New plot title", subtitle = "A subtitle")

# The caption appears in the bottom-right, and is often used for
# sources, notes or copyright
p + labs(caption = "(based on data from ...)")

# The plot tag appears at the top-left, and is typically used
# for labelling a subplot with a letter.
p + labs(title = "title", tag = "A")

# If you want to remove a label, set it to NULL.
p + labs(title = "title") + labs(title = NULL)
```

## <xnum>3.5</xnum> The Grammer of Graphics 

Wickham summarizes his grammer of graphics into seven essential parameters, and gives the following template for understanding ggplot2 in terms of this grammer:

    ggplot(data = <DATA>) + 
      <GEOM_FUNCTION>(
         mapping = aes(<MAPPINGS>),
         stat = <STAT>, 
         position = <POSITION>
      ) +
      <COORDINATE_FUNCTION> +
      <FACET_FUNCTION>


Note that

* ggplot2 will provide useful defaults for everything except the data, the mappings, and the geom function
* you are encouraged to play around with different options from a basic plot. Note that you can store a basic plot object in a variable and add features to it incrementally.
* there is also [a *qplot* function](https://ggplot2.tidyverse.org/reference/translate_qplot_ggplot.html) that makes even more default assumptions. It can be helpful to see qhat choices qplot makes.
